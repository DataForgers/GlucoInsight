```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:,1]

    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

    print("ROC AUC Score:", roc_auc_score(y_test, y_prob))
    print("Accuracy:", accuracy_score(y_test, y_pred))

    return {
        'classification_report': classification_report(y_test, y_pred, output_dict=True),
        'roc_auc': roc_auc_score(y_test, y_prob),
        'accuracy': accuracy_score(y_test, y_pred)
    }
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, precision_recall_curve

smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

print(f"Original dataset shape:\n{y.value_counts()}")
print(f"Resampled dataset shape:\n{y_res.value_counts()}")

# Split the resampled data
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train a Random Forest Classifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)

# Predictions
y_pred_rf = rf.predict(X_test_scaled)
y_prob_rf = rf.predict_proba(X_test_scaled)[:, 1]

# Evaluate Random Forest Model
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))

# Define the parameter grid for XGBoost
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'scale_pos_weight': [1, 3, 5]  
}

# Initialize GridSearchCV
xgb_grid = GridSearchCV(XGBClassifier(random_state=42), param_grid,
                        cv=3, scoring='roc_auc', n_jobs=-1, verbose=2)

# Fit on training data
xgb_grid.fit(X_train_scaled, y_train)

# Get the best model
best_xgb = xgb_grid.best_estimator_
print("\nBest XGBoost Parameters:", xgb_grid.best_params_)

# Get probabilities from best XGBoost model
y_probs = best_xgb.predict_proba(X_test_scaled)[:, 1]

# Compute precision-recall curve
precisions, recalls, thresholds = precision_recall_curve(y_test, y_probs)

# Handle case where there are no thresholds returned
if len(thresholds) > 0:
    f1_scores = (2 * precisions[:-1] * recalls[:-1]) / (precisions[:-1] + recalls[:-1] + 1e-9)
    best_threshold = thresholds[np.argmax(f1_scores)]
else:
    best_threshold = 0.5  

print("\nBest Threshold:", best_threshold)

# Apply new threshold
y_pred_adjusted = (y_probs >= best_threshold).astype(int)

# Define evaluation function
def evaluate_model(y_true, y_pred, y_prob):
    metrics = {
        "Confusion Matrix": confusion_matrix(y_true, y_pred),
        "ROC AUC Score": roc_auc_score(y_true, y_prob),
        "Classification Report": classification_report(y_true, y_pred)
    }
    return metrics

# Call evaluation function
print("\nFinal XGBoost Results:")
final_metrics = evaluate_model(y_test, y_pred_adjusted, y_probs)

# Print final evaluation metrics
print("\nFinal Metrics:")
for key, value in final_metrics.items():
    print(f"\n{key}:\n{value}")
```

```python
import joblib
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, f1_score
from imblearn.under_sampling import RandomUnderSampler

# Defining selected features and target
features = ["HighBP", "GenHlth", "NoDocbcCost", "Age", "BMI", "Smoker", "Sex", "Income", "Education", "PhysActivity"]
X = df[features]
y = df["Diabetes_012"]

# Print original class distribution
print("Original dataset shape:\n", y.value_counts())

# Apply SMOTE for oversampling minority classes
smote = SMOTE(sampling_strategy="not majority", random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Apply undersampling to balance majority class
under_sampler = RandomUnderSampler(random_state=42)
X_res, y_res = under_sampler.fit_resample(X_res, y_res)

# Print resampled class distribution
print("\nResampled dataset shape:\n", y_res.value_counts())

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[features])  
X_test_scaled = scaler.transform(X_test[features])

# Save the scaler for deployment
joblib.dump(scaler, "scaler2.pkl")

# Train a Random Forest Classifier as a baseline
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train_scaled, y_train)

# Predictions with Random Forest
y_pred_rf = rf.predict(X_test_scaled)
y_prob_rf = rf.predict_proba(X_test_scaled)[:, 1]

# Evaluate Random Forest Model
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))

# Define the parameter grid for XGBoost
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [3, 5],
    "learning_rate": [0.01, 0.1],
    "scale_pos_weight": [1, 3],
}

# Initialize GridSearchCV for XGBoost
xgb_grid = GridSearchCV(
    XGBClassifier(random_state=42, use_label_encoder=False, eval_metric="logloss"),
    param_grid,
    cv=3,
    scoring="roc_auc",
    n_jobs=-1,
    verbose=2,
)

# Fit on training data
xgb_grid.fit(X_train_scaled, y_train)

# Get the best model
best_xgb = xgb_grid.best_estimator_
print("\nBest XGBoost Parameters:", xgb_grid.best_params_)

# Make probability predictions
y_probs = best_xgb.predict_proba(X_test_scaled)[:, 1]

# Find the best threshold for classification
thresholds = np.arange(0.1, 0.9, 0.01)  
best_f1 = 0
best_threshold = 0.5

for threshold in thresholds:
    y_pred_temp = (y_probs >= threshold).astype(int)
    f1 = f1_score(y_test, y_pred_temp, average="weighted")
    if f1 > best_f1:
        best_f1 = f1
        best_threshold = threshold

print(f"\nBest Threshold Found: {best_threshold:.2f}")

# Make final predictions using the best threshold
y_pred_adjusted = (y_probs >= best_threshold).astype(int)

# Evaluate model performance
print("\nFinal XGBoost Classification Report with Adjusted Threshold:")
print(classification_report(y_test, y_pred_adjusted))

# Print confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_adjusted))

# Print ROC AUC Score
print("\nROC AUC Score:", roc_auc_score(y_test, y_probs))

# Save the best model for deployment
joblib.dump(best_xgb, "diabetes_xgb_model.pkl")

# Feature Importance Analysis
importances = best_xgb.feature_importances_
sorted_indices = np.argsort(importances)[::-1]

print("\nFeature Importances:")
for i in sorted_indices[:10]:  
    print(f"{features[i]}: {importances[i]:.4f}")

# Plot Feature Importance
plt.figure(figsize=(10, 5))
sns.barplot(x=importances[sorted_indices[:10]], y=[features[i] for i in sorted_indices[:10]])
plt.title("Top 10 Important Features")
plt.xlabel("Feature Importance Score")
plt.ylabel("Features")
plt.show()
```
